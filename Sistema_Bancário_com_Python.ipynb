# **REGRAS**

**Deposito**

Exibir mensagem "Qual o valor que deseja depositar?" (valor inteiro e positivo);

As operações de deposito deve aparecer no extrato.

**Saque**

3 saques diários com limite máximo de R$500 (por saque);

Se não tiver saldo em conta, exibir mensagem “Não é possível sacar o dinheiro pois não tem saldo.”;

Todos os saques devem aparecer no extrato.

**Extrato**

Deve listar todos os saques e depósitos realizados na conta;

No fim da listagem deve exibir o valor atual da conta;

Os valores devem ser exibidos: R$ xxx.xx (exemplo: R$ 1500.45).


```python
menu = """
[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:
  opcao = input(menu)

  if opcao == "d":
    valor = float(input("Digite o valor para deposito: "))
    if valor < 0:
      print("Operação inválida")
    else:
      print("Depósito realizado com sucesso!")
      saldo += valor
      extrato += f"Depósito: R$ {valor:.2f}\n"

  elif opcao == "s":
    valor = float(input("Digite o valor para saque: "))

    if valor > saldo:
      print("Não é possível sacar o dinheiro pois não tem saldo suficiente")
    elif valor > limite:
      print("Você excedeu o limite de saque diário")
    elif numero_saques >= LIMITE_SAQUES:
      print("Você excedeu o número de saques")
    elif valor > 0:
      print("Saque realizado com sucesso!")
      saldo -= valor
      extrato += f"Saque: R$ {valor:.2f}\n"
      numero_saques += 1
    else:
      print("Operação inválida")


  elif opcao == "e":
    print("\nEXTRATO\n")
    if not extrato:
      print("Não foram efetuadas movimentações.")
    else:
      print(extrato)
      print(f"Saldo: R$ {saldo:.2f}\n")

  elif opcao == "q":
    break
  else:
        print("Operação invalida, por favor, selecione a operação desejada.")

```
